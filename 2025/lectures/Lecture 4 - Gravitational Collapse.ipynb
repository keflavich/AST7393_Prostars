## Pressureless Collapse: the Accretion Rate

Starting from a uniform density medium, how does collapse progress?  (ch 6.3)

How fast does the newborn protostar gain (accrete) mass?  (i.e., what is the accretion rate?)

The mass of a spherically symmetric object in spherical coordinates is given by:

\begin{equation}
M_r =  \int_0^r 4\pi r'^2 \rho(r') \, dr'
\end{equation}


or equivalently
\begin{equation}
\frac{\partial M_r}{\partial r} = 4\pi r^2 \rho.
\end{equation}


The equation of conservation of mass for the gas in spherical coordinates is
<div class=fragment>
\begin{eqnarray}
\frac{\partial}{\partial t} \rho + \nabla\cdot (\rho \vec{v}) & = & 0 \\
\end{eqnarray}
</div>
<div class=fragment>
\begin{eqnarray}
\frac{\partial}{\partial t} \rho + \frac{1}{r^2}\frac{\partial}{\partial r}(r^2 \rho v) & = & 0
\end{eqnarray}
</div>
<div class=fragment>
where $v$ is the radial velocity of the gas, and the divergence in spherical coordinates is $\nabla\cdot x=\frac{1}{r^2}\frac{\partial (r^2 x)}{\partial r}$ 
</div>
<div class=fragment>
$\frac{\partial \rho}{\partial t}$ is the change of density at any point.
$\nabla\cdot (\rho \vec{v})$ is the momentum density gradient, or the rate at which material is flowing through.
    </div>

The Lagrangian version of the momentum equation (following fluid elements, not locations in space) is

$$ \rho \frac{Dv}{Dt} = - \frac{\partial}{\partial r} P - \rho \mathbf{f}_g$$

<div class=fragment>
i.e., the acceleration of a particle is set by the sum of pressure and gravity.
</div>

<div class=fragment>
Assuming there is no thermal pressure (i.e., $c_s=0$, so $P=0$), it becomes:
\begin{equation}
\frac{Dv}{Dt} = -\frac{GM_r}{r^2}.
\end{equation}
    </div>
    
<div class=fragment>
This just says that a shell's inward acceleration is equal to the gravitational force per unit mass exerted by all the mass interior to it, which is constant (as long as the inside is collapsing faster than the outside). 
    </div>

We solve this by changing variables:
$$
\frac{Dv}{Dt} = \frac{dv}{dr}\frac{dr}{dt} = v \frac{dv}{dr} = - \frac{GM_r}{r^2}
$$

<div class=fragment>
<div>
We can then integrate:

<div>
$$
v \frac{dv}{dr} = - \frac{GM_r}{r^2}
$$</div>
<div class=fragment>
$$ \int_{v_0}^V v' dv' = \int_{r_0}^r -\frac{GM_r}{r'^2} dr' $$
</div>

<div  class=fragment>
    $$\left[\frac{1}{2}v'^2\right]_{v_0,V} = \left[\frac{G M_r}{r'}\right]_{r_0,r}$$
</div>
    
<div  class=fragment>
    $$\frac{1}{2}(V^2-v_0^2) = G M_r \left(\frac{1}{r} - \frac{1}{r_0}\right)$$
</div>
</div>   
</div>   

We have

$$\frac{1}{2}(V^2-v_0^2) = G M_r \left(\frac{1}{r} - \frac{1}{r_0}\right)$$

<div class=fragment>
If we start from $v_0=0$ (at rest), but $r_0$ is finite, 

$$\frac{1}{2} V^2 = G M_r \left(\frac{1}{r} - \frac{1}{r_0}\right)$$

</div>
<div class=fragment>
becomes

$$ v = \sqrt{2 G M_r} \left(\frac{1}{r} - \frac{1}{r_0}\right)^{1/2}$$
</div>

We have the solution for the velocity as a function of position:
\begin{equation}
v = \dot{r} = -\sqrt{2GM_r}\left(\frac{1}{r}-\frac{1}{r_0}\right)^{1/2},
\end{equation}
where $r_0$ is the position at which a particular fluid element starts. 


This is the infall velocity as a function of position, so it tells you how fast material is flying to the center at each spot.

Aside:  Why is pressureless collapse a good assumption?
"Consider the virial theorem: the thermal pressure
term is just proportional to the mass, since the gas sound speed stays
about constant. On the other hand, the gravitational term varies as
1/R. Thus, even if pressure starts out competitive with gravity, as the
core collapses the dominance of gravity will increase, and before too
long the collapse will resemble a pressureless one."

This leads to an integral for the position of our fluid element (our infalling blob of gas) with time:
\begin{equation}
r(t) = \int -\sqrt{2GM_r}\left(\frac{1}{r}-\frac{1}{r_0}\right)^{1/2} dt
\end{equation}



The solution is (using some trig cleverness spelled out in Krumholz), with $r = r_0 \cos^2 \xi$
\begin{equation}
\xi+\frac{1}{2}\sin 2\xi  =  t \sqrt{\frac{2GM_r}{r_0^3}}
\end{equation}



We want to know the time $t$ at which $r=0$, i.e., where the outermost layer has collapsed into the center.  This occurs at $\cos \xi = 0$ or $\xi=\pi/2$.

With $\xi=\pi/2$, 
\begin{equation}
\xi+\frac{1}{2}\sin 2\xi  =  t \sqrt{\frac{2GM_r}{r_0^3}}
\end{equation}


becomes, recognizing that $\sin 2 \xi = \sin \pi = 0$ and rearranging,
\begin{equation}
t = \frac{\pi}{2} \sqrt{\frac{r_0^3}{2GM_r}} = \sqrt{\frac{\pi^2 r_0^3}{8GM_r}}
\end{equation}



For a uniform density $M_r = 4/3 \pi \rho_0 r_0^3$, so
\begin{equation}
t_{ff} = \sqrt{\frac{3\pi}{32 G \rho_0}}
\end{equation}

For uniform density, 
\begin{equation}
t_{ff} = \sqrt{\frac{3\pi}{32 G \rho_0}}
\end{equation}
holds for all fluid elements: everything reaching the center gets there at the same time!

For a more general case, in which the core is centrally condensed with central density $\rho_c$ and a power-law density profile $\rho(r) = \rho_c (r/r_c)^{-\alpha} $, the enclosed mass is instead:

\begin{equation}
M_r = \frac{4}{3-\alpha}\pi \rho_c r_c^3 \left(\frac{r}{r_c}\right)^{3-\alpha} 
\end{equation}



which can be substituted into $t =  \sqrt{\frac{\pi^2 r_0^3}{8GM_r}}$
 to obtain a collapse timescale:

\begin{equation}
t_{ff} = \sqrt{\frac{(3-\alpha)\pi }{32 G \rho_c }} \left(\frac{r_0}{r_c}\right)^{\alpha/2}
\end{equation}

\begin{eqnarray}
M_r &=& \int_0^r 4 \pi r'^2 \rho(r') dr' \\
    &=& \int_0^r 4 \pi r'^2 \rho_0 \left(\frac{r'}{r_c}\right)^{-\alpha} dr' \\
    &=& 4 \pi \rho_0 r_c^{-\alpha} \int_0^r \left(r'\right)^{2-\alpha}  dr' \\
    &=& 4 \pi \rho_0 r_c^{-\alpha} \frac{1}{3-\alpha} \left[\left(r'\right)^{3-\alpha}\right]_0^r   \\
    &=& \frac{4}{3-\alpha}\pi \rho_c r_c^3 \left(\frac{r}{r_c}\right)^{3-\alpha} 
\end{eqnarray}

In this general case, with $\alpha>0$ (required for a centrally condensed object):
\begin{equation}
t_{ff} = \sqrt{\frac{(3-\alpha)\pi }{32 G \rho_c }} \left(\frac{r_0}{r_c}\right)^{\alpha/2}
\end{equation}
the free-fall time increases with radius, $t_{ff} \propto r_0^{\alpha/2}$.

That means the innermost parts of the object collapse first.

*Inside-out collapse* is a key feature of pressureless collapse.

We now have $\frac{\partial}{\partial t} M_r = -v \frac{\partial}{\partial r}M_r$, a differential equation relating the time evolution of mass (the accretion rate) to the structure of the core.

\begin{equation}
t_{ff} = \sqrt{\frac{3\pi}{32 G \rho_0}}
\end{equation}

What is the free-fall time for $n(H_2) = 10^4\mathrm{~cm^{-3}}$?

### The shape of a collapsing core

We'll next derive the shape (i.e., the density profile $\rho(r)$) of a collapsing core.



To determine the shape of a core, it is useful to write the mass conservation equation in terms of $M_r$ rather than $\rho$.
<table width=100%><tr>
<td> We start with $M_r$:
\begin{equation}
M_r =  \int_0^r 4\pi r'^2 \rho(r') \, dr'
\end{equation}
        and take the time derivative $\frac{\partial}{\partial t}$
</td>
<td class=fragment>
\begin{eqnarray}
\frac{\partial}{\partial t}M_r & = & 4\pi \int_{0}^{r} r'^2 \frac{\partial}{\partial t} \rho \,dr' \\
& = & -4\pi \int_{0}^{r} \frac{\partial}{\partial r'}(r'^2 \rho v)\, dr' \\
& = & -4\pi r^2 \rho v \\
& = & -v \frac{\partial}{\partial r}M_r
\end{eqnarray}
</td></tr>
<tr ><td colspan=2 class=fragment>
    In the second step we used the mass conservation equation to substitute for $\partial \rho/\partial t$:
    \begin{equation}
    \frac{\partial}{\partial t} \rho = - \frac{1}{r^2}\frac{\partial}{\partial r}(r^2 \rho v) 
    \end{equation}
</td>
</tr>
<tr><td colspan=2 class=fragment>
        In the final step we used the definition of $M_r$:
$\frac{\partial M_r}{\partial r} = 4\pi r^2 \rho$ 
        to substitute for $\rho$.
</td></tr>
</table>

The mass conservation equation is now
\begin{equation}
\frac{\partial M_r}{\partial t} = -v\frac{\partial M_r}{\partial r}  = -4\pi r^2 v \rho
\end{equation}


Solving the mass conservation equation
\begin{equation}
\dot{M} \equiv \frac{\partial M_r}{\partial t} = -v\frac{\partial M_r}{\partial r}  = -4\pi r^2 v \rho
\end{equation}
<div class=fragment>
for density $\rho$, and using $v=v_{ff}=- \sqrt{\frac{2 G M_r}{r}}$ gives:

</div>
<div class=fragment>

\begin{equation}
\rho = \frac{(\partial M_r/\partial t) r^{-3/2}}{4\pi\sqrt{2 G M_r}} = \frac{\dot{M} }{4\pi\sqrt{2 G M_r}} r^{-3/2}
\end{equation}
</div>
<div class=fragment>
Under the assumption that $\dot{M}$ (the _accretion rate_) is constant, that means the density in the inner portion of a freely collapsing core will tend to 

\begin{equation}
 \rho \propto r^{-3/2}
\end{equation}
</div>

Explanation of $v_{ff}$ in cell above:

For a free-falling fluid element, the speed will approach the free-fall velocity, i.e.:

\begin{equation}
 v = v_{ff} = - \sqrt{\frac{2 G M_r}{r}}
\end{equation}

### Accretion Rate

What is the time-averaged accretion rate (for a centrally-condensed core)?

<div class=fragment>
We can plug in $r=r_c$ to get the free-fall time for a fluid element from radius $r_c$:
\begin{equation}
M_c = [4/(3-\alpha)]\pi \rho_c r_c^3
\end{equation}

\begin{equation}
t_c = \sqrt{\frac{(3-\alpha)\pi}{32 G \rho_c}} = \sqrt{\frac{3-\alpha}{3}}t_{\rm ff}(\rho_c)
\end{equation}

where $t_{\rm ff}(\rho_c)$ is the free-fall time for an object with constant density $\rho_c$
</div>

The average accretion rate is 
\begin{equation}
\left<\dot{M}\right> = \frac{M_c}{t_c} = \sqrt{\frac{3}{3-\alpha}} \frac{M_c}{t_{\rm ff}(\rho_c)}
\end{equation}

We need a mass to evaluate the accretion rate.  For a critical Bonnor-Ebert sphere (the largest mass that can be gravitationally stable in the presence of external pressure), we have

\begin{equation}
M_{\rm BE} = 1.18 \frac{c_s^4}{\sqrt{G^3 P_s}},
\end{equation}

where $c_s$ is the sound speed, and $P_s$ is the pressure at the surface of the sphere.

If we assume the sphere is in pressure equilibrium with its surroundings, the internal pressure $\rho c_s^2$ equals the outer $P_S=\rho c_s^2$, so 
\begin{equation}
M_{\rm BE} = 1.18 \frac{c_s^4}{\sqrt{G^3 P_s}},
\end{equation}
becomes
\begin{equation}
M_{\rm BE} = 1.18 \frac{c_s^4}{\sqrt{G^3 \rho_c c_s^2}} = 1.18 \frac{c_s^3}{\sqrt{G^3 \rho_c}},
\end{equation}




Then, the accretion rate becomes:
\begin{equation}
\left<\dot{M}\right> = \sqrt{\frac{3}{3-\alpha}} \frac{M_{BE}}{t_{\rm ff}(\rho_c)} = 1.18 \sqrt{\frac{3}{3-\alpha}} \frac{c_s^3 \sqrt{32 G \rho_c}}{\sqrt{G^3 \rho_c3 \pi}} = \frac{c_s^3}{G}\frac{\sqrt{32} \cdot 1.18}{\sqrt{(3-\alpha) \pi}}
\end{equation}

The accretion rate is:
\begin{equation}
\left<\dot{M}\right>  \approx \frac{c_s^3}{G}
\end{equation}

i.e., it's independent of the initial density.

## The Singular Isothermal Sphere

The only fully analytic solution for collapse of a gas core is the Singular Isothermal Sphere.  

There is a self-similar solution by [Shu 1977](https://ui.adsabs.harvard.edu/abs/1977ApJ...214..488S/abstract) that is important to be aware of because it's the common starting point for protostellar evolution models.



[Foster & Chevalier 1993](https://ui.adsabs.harvard.edu/abs/1993ApJ...416..303F/abstract) obtain numerical solutions for the time evolution of a (nearly) singular isothermal sphere, which has some substantial differences from the self-similar SIS solution; in particular, "realistic" cores do not have constant $\dot{M}$.

In the above slide, $\xi=\frac{r}{r_0}$.

Top-left: The infall starts at the outside; the furthest-out rings collapse the fastest (because they have the furthest to go; they all need to arrive at the center at the same time).  When v=0, steady state is reached.

Top-right: D is the density; we start with a centrally-flattened core that approaches the free-fall $\rho\propto r^{-3/2}$ solution.

Bottom-left: Where is the mass as a function of radius?  Mass is the integral $M=\int \rho 4 \pi r^2 dr$, so at the start, there's very little mass in the center, but as time goes on, the mass in the center builds up.

## Unrealistic but useful

While the analytic solutions are not realistic, since they leave out key physics, they are still useful and are used in may contexts.

For example, many of the turbulent fragmentation models invoke the Jeans mass and subsequent SIS evolution to model stars "collapsing out" of the turbulent flow.
