## Birthline

The "birthline" is where a star would appear on the HR diagram if it suddenly stopped accreting.

To construct the birthline, we assume that while a star is accreting, it is completely (optically) invisible.  Then, suddenly, it stops accreting and is optically visible. 

"The birthline is simply the locus in the HR diagram of pre-main-sequence stars with protostellar radii."  Palla & Stahler S16.1.2

The birthline is limited to a narrow range even if there is a wide range in accretion rates because of the thermostatting effect of deuterium burning: stars end up at close to the same radius for a moderately wide range of accretion rates.

## Hayashi Track

Once a star hits the birthline (stops accreting), it follows a well-defined track to the main sequence.

For low-mass stars, below $M\lesssim1 \mathrm{M}_\odot$, this evolution is primarily along the _Hayashi track_, which is vertical in the HR diagram

```python
Image('graphics/krumholzfig17.3_hayashitrack.png')
```

Key features of the Hayashi tracks:

 * Low-mass stars evolve vertically: constant temperature

Qualitative desription of why the Hayashi track is vertical:

At $T<10^4$ K, H is neutral, but most metals are ionized.  There are therefore plentiful free electrons.  These conditions allow for abundant formation of H$^-$, i.e., negatively-charged hydrogen.

H$^{-}$ has very high opacity, and its abundance depends strongly on temperature.

At low temperature, the opacity of neutral gas gets very low.

Low optical depth implies that the coupling to radiation is very weak.  But the star is held up in pressure equilibrium because the energy cannot radiate away: it is optically thick.  If the star was fully optically thin, it would radiate away its energy efficiently and contract, and then would heat back up, driving up the coupling again.

This is a "thermostatting" effect.
<!--(this argument is my own reading between the lines a bit - it's not directly from a textbook and so it could be wrong)-->

The minimum temperature for a star to have an H$^-$ photosphere is $T_{Hayashi} = 3500$ K.  So most stars stack up at this temperature.

Aside: S17.3.2 of Krumholz, or 16.2 of Palla & Stahler, goes through a quantitative explanation of the Hayashi track

## Henyey track

For more massive stars, they eventually contract enough that their surface temperature exceeds $T>T_{Hayashi}$.

At this stage, the opacity drop with increasing temperature (because H$^-$ dominates opacity) results in a change from convective to radiative energy transport.


This results in a "vaguely horizontal" track that continues until the star has contracted enough to initiate hydrogen burning, then it's on the main sequence.

Stars with $M>0.4 \mathrm{M}_\odot$ become radiative before reaching the main sequence

# Definition of Age is ambiguous.

The Zero-Age Main Sequence (ZAMS) is the curve highlighting where stars _start_ their lives from a stellar evolutionary perspective.



Core hydrogen burning begins before the ZAMS for low-mass stars - the ZAMS draws where stars have lost their gravitational energy

The _birthline_ is where a star stops accreting.  But "stops accreting" is itself ambiguous - how low does it have to go?  What about late-time accretion bursts?

There is no well-defined, clear zero-age.  

This ambiguity is a problem for discussions of star formation and ages.

# H-R diagram comparison to determine underlying physics

In the previous lecture, we noted that there are no first-principles calculatations of how "hot" or "cold" accretion is.

We can only infer how accretion happens by comparing observations of post-birthline stars to models.

[Kunimoto+ 2017](https://ui.adsabs.harvard.edu/abs/2017A%26A...599A..49K/abstract) examined the effects of hot and cold accretion and varying deuterium abundance on the HR diagram.

# Lithium Burning
Palla & Stahler 16.3.2

Lithium burns after deuterium, but before hydrogen.  $T_{Li}\sim3\times10^6$ K

Lithium is very low-abundance, $[\mathrm{Li/H}]\approx2\times10^{-9}$: it is only a trace element and not relevant for computing stellar structure.

However, it is readily observable in stellar atmospheres.

Protostars are too cool to burn lithium throughout their mass.

They burn lithium in their cores, but this burning occurs after the star becomes radiative.

Therefore, the outer parts of the star retain ISM levels of lithium.

Pre-main-sequence stars, and eventually main sequence low-mass stars, are convective.  $(M < 1.2 M_\odot$ have deep enough convective layers that photospheric materials are mixed into the lithium-burning zone)

Therefore, once they reach the birthline, the stars begin to lose their lithium.

This provides us with the best tool we have to measure stellar ages.
